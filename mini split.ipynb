```python
base_path = '/content/drive/MyDrive/DeeptraceReward/mini_splits'
json_path = os.path.join(base_path, 'labels.json')

with open(json_path, 'r') as f:
    labels = json.load(f)

new_labels = {}
for key, value in labels.items():
  slash = key.find('/')
  file_name = key[slash+1:]
  new_labels[file_name] = value

with open(json_path, 'w') as f:
    json.dump(new_labels, f, indent=4)

```

```python
base_path = '/content/drive/MyDrive/DeeptraceReward/mini_splits'
folders = ['train', 'val', 'test']

json_path = '/content/drive/MyDrive/DeeptraceReward/mini_splits/labels.json'

with open(json_path, 'r') as f:
  labels = json.load(f)

for folder in folders:
  X_list = [] # make a X-y matrix for each split
  y_list = []
  folder_path = os.path.join(base_path, folder)
  files = [f for f in os.listdir(folder_path) if f.endswith('.npy') and not f.startswith('.')]

  for f in files:
    path = os.path.join(folder_path, f)
    arr = np.load(path)
    X_list.append(arr)
    y_list.append(labels[f])

  X = np.stack(X_list, axis=0)             # shape: (n_samples, n_features)
  y = np.array(y_list).reshape(-1, 1)      # shape: (n_samples, 1)
  dataset = np.hstack((X, y))

  save_path = '/content/drive/MyDrive/DeeptraceReward/mini_splits/' + folder + '.npy'

  np.save(save_path, dataset)



```
